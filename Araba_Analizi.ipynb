```python
y=dataFrame["price"].values
x=dataFrame.drop("price",axis=1).values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #yapay sinir ağı kullan
from tensorflow.keras.layers import Dense #yapay sinir ağında katman oluştur
```

```python
x_train.shape
```

```python
model= Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train, y = y_train, validation_data=(x_test,y_test),epochs=300)
#,batch_size=240  ,validation_data=(x_test,y_test)
#validation data  = doğrulama verisini direk model içinde verdim
#batch_size=250  verim büyük ve bu  yapımı yorar o yüzden parca parca veriyorum
```
